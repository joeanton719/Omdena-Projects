```python
import pandas as pd
```

```python
folder_path="datasets/05_tr_football_matches/"

df1 = pd.read_csv(folder_path+"ist_uefa_matches.csv")
df2 = pd.read_csv(folder_path+"istanbul_football_superleague.csv")
```

```python
df2 = (
    df2.assign(league="Spor Toto Super Lig",
               stadium=lambda x:x['stadium'].replace(r"\s+-\s+[İI]STANBUL\s*$", "", regex=True).str.strip())
)
```

```python
stad_dict = (
    df2[['stadium', 'stad_lat', 'stad_long', 'stad_address']]
    .drop_duplicates()
    .set_index("stadium").T
    .to_dict()
)

stad_dict
```

```python
df2['stadium'].unique()
```

```python
df1['stadium']=\
df1['stadium'].replace(
    ['Başakşehir Fatih Terim Stadı', 
     'Türk Telekom Arena',
     'Vodafone Arena (Beşiktaş Jimnastik Kulübü İnönü Stadyumu)',
     'Fenerbahçe Şükrü Saracoğlu Stadyumu',
     'Recep Tayyip Erdoğan Stadyumu',
     'Ali Sami Yen Spor Kompleksi – Nef Stadyumu'],
     ['BAŞAKŞEHİR FATİH TERİM',
      'NEF STADYUMU',
      'VODAFONE PARK',
      'ÜLKER STADYUMU',
      'RECEP TAYYİP ERDOĞAN STADYUMU',
      'NEF STADYUMU'
      ]
)
```

```python
df = pd.concat([df1, df2]).sort_values('date', ignore_index=True)
```

```python
for stad in df1['stadium'].unique():
    for col in ['stad_lat', 'stad_long', 'stad_address']:
        df.loc[df['league'].str.startswith('UEFA') & df['stadium'].eq(stad), col] = stad_dict[stad][col]
```

```python
(
    df[['stadium', 'stad_lat', 'stad_long']]
    .sort_values("stadium")
    .drop_duplicates(ignore_index=True)
)
```
