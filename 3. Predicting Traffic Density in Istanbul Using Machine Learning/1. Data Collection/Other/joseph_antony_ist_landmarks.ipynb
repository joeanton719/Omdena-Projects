```python
import pandas as pd
from geopy.geocoders import Nominatim
from tqdm.auto import tqdm
```

```python
sites = [
    "Hagia Sophia",
    "Topkapı Palace",
    "Kariye Museum",
    "Süleymaniye Mosque",
    "Basilica Cistern",
    "Blue Mosque",
    "Kapalı Çarşı",
    "Pera Museum",
    "İstanbul Arkeoloji Müzeleri",
    "Dolmabahçe Palace",
    "Museum of Innocence",
    "Museum of Turkish & Islamic Arts",
    "Rahmi M Koç Museum",
    "Egyptian Spice Bazaar",
    "İstanbul Modern Museum",
    "İstiklal Caddesi",
    "Beylerbeyi Palace",
    "Galata Tower",
    "Galata Bridge",
    "Taksim Square - Istanbul",
    "Gülhane Parkı",
    "Meryem Ana Rum Ortodoks Kilisesi - Balat",
    "Pierre Loti - Eyüp Merkez",
    "Eyüp Sultan Mosque",
    "Küçük Ayasofya",
    "500. Yıl Vakfı Türk Musevileri Müzesi",
    "Rumeli Hisarı",
    "Atik Valide Mosque",
    "Galata Mevlevihanesi Müzesi",
    "Valens Su Kemeri",
    "Deniz Müzesi",
    "Çırağan Palace",
    "Harbiye Askeri Müze",
    "Yıldız Park - Beşiktaş",
    "Ortaköy Büyük Mecidiye Mosque",
    "5 Temmuz Şehitler Köprüsü",
    "Fatih Sultan Mehmet Köprüsü",
    "Yavuz Sultan Selim Bridge",
    "Çamlıca Kulesi - Istanbul",
    "Sabiha Gökçen International Airport",
    "İstanbul Atatürk Havalimanı",
    "İstanbul Havalimanı"
]
```

```python
geolocator = Nominatim(user_agent="ist_sites")
```

```python
sites_info=[]
for site in tqdm(sites):
    
    location = geolocator.geocode(site)

    if location:
        sites_info.append({
            "site_name" : site,
            "latitude" : location.latitude,
            "longitude" : location.longitude,
            "site_address" : location.address
        })


    else:
        print(site)
```

```python
df = (
    pd.concat([
        pd.DataFrame(sites_info),
        pd.DataFrame([{"site_name": "İBB 15 Temmuz Şehitler Ormanı ve Anıtı",
                      "latitude" : 41.2075602,
                      "longitude" : 29.0958613,
                      "site_address" : "Garipçe Mahallesi, Sarıyer, İstanbul, Marmara Bölgesi, 34450, Türkiye"}])
    ])
    .sort_values('site_name')
)

df
```
